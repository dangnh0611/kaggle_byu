```python
import numpy as np

all_target_z = []
all_target_y = []
all_target_x = []

SAVE_VIZ_DIR = '/home/dangnh36/datasets/.comp/byu/processed/viz/external_annotations_per_dataset/'

os.makedirs(SAVE_VIZ_DIR, exist_ok=True)

for i, row in tqdm(label_df.iterrows()):
    tomo_id = row['tomo_id']
    if tomo_id == 'mba2011-07-18-1':
        all_target_z.append(-1)
        all_target_y.append(-1)
        all_target_x.append(-1)
        continue
    Z, Y, X = eval(row['target_shape'])
    ori_Z, ori_Y, ori_X = eval(row['ori_shape'])
    z_indices = np.linspace(0, ori_Z-1, 128).astype(int)
    assert len(z_indices) == 128
    z = row['z']
    assert z == int(z)
    z = int(z)
    ori_z = z_indices[z]
    target_z = (ori_z + 0.5) * Z / ori_Z - 0.5
    target_y = (row['y'] + 0.5) / 512. * Y - 0.5
    target_x = (row['x'] + 0.5) / 512. * X - 0.5
    all_target_z.append(target_z)
    all_target_y.append(target_y)
    all_target_x.append(target_x)

    img = cv2.imread(os.path.join(TOMO_DIR, tomo_id, f'slice_{round(target_z):04d}.jpg'))
    radius = round(1000. / row['target_spacing'])
    cv2.circle(
        img,                    # Image
        center=(round(target_x), round(target_y)),      # (x, y) center coordinates
        radius=radius,              # Radius of the circle
        color=(0, 255, 0),      # Color in BGR (green)
        thickness=2             # Thickness (-1 to fill the circle)
    )
    save_path = os.path.join(SAVE_VIZ_DIR, str(row['dataset_id']), f'{i}__{tomo_id}.jpg')
    os.makedirs(os.path.dirname(save_path), exist_ok = True)
    cv2.imwrite(save_path, img)
```
