```python
import polars as pl
import json
import sys
import sklearn
from matplotlib import pyplot as plt
```

```python
df = pl.scan_csv('/home/dangnh36/datasets/.comp/byu/processed/gt_v2.csv').sort('tomo_id').collect()
df
```

```python
NUM_FOLDS = 4
RANDOM_STATE = 42
all_tomo_ids = df['tomo_id'].unique().to_list()
print('TOTAL NUM TOMOS:', len(all_tomo_ids))


splitter = sklearn.model_selection.StratifiedKFold(n_splits=NUM_FOLDS, shuffle=True, random_state=RANDOM_STATE)

cv_ret = {
    'splitter': 'sklearn.model_selection.StratifiedKFold',
    'num_folds': NUM_FOLDS,
    'random_state': RANDOM_STATE,
    'folds': []
}
for i, (train_indices, val_indices) in enumerate(splitter.split(df['tomo_id'], df['num_motors'])):
    print(f"Fold {i}:")
    val_tomo_ids = df[val_indices].filter(pl.col('num_motors') <= 1)['tomo_id'].to_list()
    assert len(val_tomo_ids) == len(set(val_tomo_ids))
    train_tomo_ids = list(set(all_tomo_ids).difference(set(val_tomo_ids)))
    print(f'train={len(train_tomo_ids)} val={len(val_tomo_ids)}')
    plt.hist( df.filter(pl.col('tomo_id').is_in(val_tomo_ids))['num_motors'], bins = 11, range=(0, 10), alpha=0.5, label = 'val')
    plt.hist( df.filter(pl.col('tomo_id').is_in(train_tomo_ids))['num_motors'], bins = 11, range=(0, 10), alpha = 0.5, label = 'train')
    plt.show()
    cv_ret['folds'].append({
        'train': train_tomo_ids,
        'val': val_tomo_ids
    })

assert len(cv_ret['folds']) == NUM_FOLDS

print(cv_ret)
with open(f'/home/dangnh36/datasets/.comp/byu/processed/cv/v2/skf{NUM_FOLDS}_rd{RANDOM_STATE}.json', 'w') as f:
    json.dump(cv_ret, f)
```

```python
import polars as pl

df = pl.read_csv('/home/dangnh36/datasets/.comp/byu/processed/all_gt_v3.csv')
df
```

```python
tomo2num = {}
for row in df.iter_rows(named = True):
    tomo2num[row['tomo_id']] = row['num_motors']
print(tomo2num)
```

```python
import json

with open(f'/home/dangnh36/datasets/.comp/byu/processed/cv/v2/skf4_rd42.json', 'r') as f:
    cv_meta = json.load(f)
print(cv_meta)
```

```python
new_meta = {}
for k in ['splitter', 'num_folds', 'random_state']:
    new_meta[k] = cv_meta[k]
new_meta['folds'] = []
for fold_idx, fold_meta in enumerate(cv_meta['folds']):
    print('FOLD NUMBER', fold_idx)
    assert len(fold_meta) == 2
    new_fold_meta = {}
    train_tomo_ids = fold_meta['train']
    val_tomo_ids = fold_meta['val']
    print('OLD:', len(train_tomo_ids), len(val_tomo_ids))
    new_val_tomo_ids = [tomo_id for tomo_id in val_tomo_ids if tomo2num[tomo_id] <= 1]
    new_train_tomo_ids = train_tomo_ids + [tomo_id for tomo_id in val_tomo_ids if tomo2num[tomo_id] > 1]
    print('NEW:', len(new_train_tomo_ids), len(new_val_tomo_ids))
    new_fold_meta['train'] = new_train_tomo_ids
    new_fold_meta['val'] = new_val_tomo_ids
    new_meta['folds'].append(new_fold_meta)
    print('------------------------\n\n')
```
