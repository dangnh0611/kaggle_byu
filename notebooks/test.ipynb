```python
import cv2
import PIL.Image
from IPython.display import Image
from IPython.display import display as idisplay
import numpy as np
import torch


def longest_resize(img, max_h = None, max_w = None, upscale = False, interpolation = cv2.INTER_LINEAR):
    if max_h is None and max_w is None:
        return img
    img_h, img_w = img.shape[:2]
    _ratios = []
    if max_h is not None:
        _ratios.append(max_h / img_h)
    if max_w is not None:
        _ratios.append(max_w / img_w)
    r = min(_ratios)
    if not upscale:
        r = min(1.0, r)
    if r == 1.0:
        return img
    new_h, new_w = int(r * img_h), int(r * img_w)
    img = cv2.resize(img, (new_w, new_h), interpolation)
    return img
    

def display_img(img, max_h = 320, max_w = 320):
    if isinstance(img, torch.Tensor):
        # img = img.permute(1,2,0).numpy()
        img = img.numpy()
    img = longest_resize(img, max_h, max_w)
    print('display:', img.shape, img.dtype)
    if img.dtype != np.uint8:
        img = (img * 255).astype(np.uint8)
    idisplay(PIL.Image.fromarray(img))
    

img = cv2.imread('/home/dangnh36/datasets/.comp/byu/processed/viz/annotations/tomo_b54396.jpg')
# img = torch.from_numpy(img).permute(2, 0, 1)
img = torch.from_numpy(img[..., 0])
print(img.shape, img.dtype)
display_img(img)
```

```python
from yagm.transforms.tta_3d import build_tta
```

```python
ALL_2D_TTAS = [
    'yx',
    'yx_x',
    'yx_y',
    'yx_xy',
    'xy',
    'xy_x',
    'xy_y',
    'xy_xy',
]

uniques = []
for tta_code in ALL_2D_TTAS:
    tta_tf = build_tta(tta_code, 1.0, 'yx')
    tf_img, _ = tta_tf.transform(img)
    back_img, _ = tta_tf.invert(tf_img)
    assert torch.sum(torch.abs(img - back_img)) == 0

    for e in uniques:
        assert e.shape != tf_img.shape or torch.sum(torch.abs(e - tf_img)) != 0
    uniques.append(tf_img)
    print(tta_code)
    display_img(tf_img)
    print('--------------------------\n')

print(len(uniques))
```
